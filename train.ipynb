```python
import pandas as pd
import numpy as np
```

```python
PATH = "./data/"
train_df = pd.read_csv(f'{PATH}/train.csv')
test_df = pd.read_csv(f'{PATH}/test.csv')
sample_df = pd.read_csv(f'{PATH}/sample_submission.csv')
```

```python
train_df = train_df.fillna(train_df.mean())
test_df = test_df.fillna(test_df.mean())
total_df = pd.concat([train_df, test_df], join='inner')
```

```python
total_df
```

```python
from sklearn.preprocessing import LabelEncoder, RobustScaler

le = LabelEncoder()

total_df['attribute_0'] = le.fit_transform(total_df['attribute_0'])
total_df['attribute_1'] = le.fit_transform(total_df['attribute_1'])
total_df['product_code'] = le.fit_transform(total_df['product_code'])
```

```python
train = total_df.iloc[:train_df.shape[0], :]
test = total_df.iloc[train_df.shape[0]:, :]
```

```python
X = train.drop("id", axis=1)
y = train_df.iloc[:, -1:]
```

```python
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras

opt = keras.optimizers.Adam(learning_rate=0.001)
model = keras.Sequential()
model.add(Dense(units=512, input_dim=X.shape[1], kernel_initializer='he_normal', activation='softsign'))
model.add(Dense(units=256, activation='softsign'))
model.add(Dense(units=1, activation='softsign'))

model.compile(loss="mean_squared_error", optimizer=opt, metrics=["accuracy"])

model.fit(X, y, batch_size=500, epochs=15)
```

```python
column_id = test["id"]
test = test.drop("id", axis=1)
```

```python
pred = model.predict(np.array(test))
prediction = pd.DataFrame(column_id, columns=["id"])
prediction['failure'] = pred
```
